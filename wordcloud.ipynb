# Картина дня в облаке тегов

```python
from myrsslib import *
import sqlite3
from wordcloud import WordCloud
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
%matplotlib inline

path_to_db = 'news.db'
conn = sqlite3.connect(path_to_db)
c = conn.cursor()

c.execute("""SELECT time, title FROM titles
            ORDER BY time DESC
            LIMIT 900""")

titles, times = [], []
for time, title in c.fetchall():
    titles.append(title)
    times.append(time)

# STOPWORDS
russian_stopwords = stopwords.words("russian")
more_stopwords = ['году','года', 'власти', 'заявили','заявил','области','рф']

# Wordcloud
wordcloud = WordCloud(max_words=50, stopwords=russian_stopwords+more_stopwords, background_color='white').generate(content)

plt.figure(figsize=(12, 18))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
